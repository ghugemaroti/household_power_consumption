# Importing libraries

```python
import numpy as np # linear algebra
import pandas as pd # data procssing, CSV file I/O (e.g. pd.read_csv),for data manipulation 
import matplotlib.pyplot as plt # this is used for the plot the graph 
import seaborn as sns # used for plot interactive graph. 
```

# Reading the dataset

```python
# Download household power consumption data from this link 
# http://archive.ics.uci.edu/ml/machine-learning-databases/00235/

df = pd.read_csv('household_power_consumption.txt', sep=';', 
                 parse_dates={'dt' : ['Date', 'Time']}, infer_datetime_format=True, 
                 low_memory=False, na_values=['nan','?'], index_col='dt')
```

# Calculating the total power consumed in by Month

```python
# Calculate Global_active_power consumed in by Month

total_Gap = df['Global_active_power'].resample('M').sum()
print(total_Gap)
```

```python
# Calculate the power of Sub_metering_1 consumed in by Month

sub_m1 = df['Sub_metering_1'].resample('M').sum()
print(sub_m1)
```

```python
# Calculate the power of Sub_metering_2 consumed in by Month

sub_m2 = df['Sub_metering_2'].resample('M').sum()
print(sub_m2)
```

```python
sub_m3 = df['Sub_metering_3'].resample('M').sum()
print(sub_m3)
```

# Data Visualization using Plot

```python
# Sum of 'Global_active_power' resampled over month

df['Global_active_power'].resample('M').sum().plot(kind='bar')
plt.xticks(rotation=90)
plt.ylabel('Global_active_power')
plt.title('Global_active_power per month')
plt.show()

# Global_active_power resampled over day for sum

df.Global_active_power.resample('D').sum().plot(title='Global_active_power resampled over day for sum',
                                                color='black') 
plt.tight_layout()
plt.show()   
```

```python
#Sub_metering_1 resampled over month for sum

df['Sub_metering_1'].resample('M').sum().plot(kind='bar')
plt.xticks(rotation=90)
plt.ylabel('Sub_metering_1')
plt.title('Sub_metering_1 per month')
plt.show()

#Sub_metering_1 resampled over day for sum

df.Sub_metering_1.resample('D').sum().plot(title='Sub_metering_1 resampled over day for sum') 
plt.tight_layout()
plt.show()   

```

```python
# Sub_metering_2 resampled over month for sum

df['Sub_metering_2'].resample('M').sum().plot(kind='bar')
plt.xticks(rotation=90)
plt.ylabel('Sub_metering_2')
plt.title('Sub_metering_2 per month')
plt.show()

# Sub_metering_2 resampled over day for sum

df.Sub_metering_2.resample('D').sum().plot(title='Sub_metering_2 resampled over day for sum') 
plt.tight_layout()
plt.show()  
```

```python
# Sub_metering_3 resampled over month for sum
df['Sub_metering_3'].resample('M').sum().plot(kind='bar')
plt.xticks(rotation=90)
plt.ylabel('Sub_metering_3')
plt.title('Sub_metering_3 per month')
plt.show()

# Sub_metering_3 resampled over day for sum

df.Sub_metering_3.resample('D').sum().plot(title='Sub_metering_3 resampled over day for sum') 
plt.tight_layout()
plt.show()  
```
